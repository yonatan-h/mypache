```python
# generate 
from random import random
xy_pairs = [] #for 
for i in range(100):
    x = i
    y = 2*x + 1 + random()*5 - 2.5
    xy_pairs.append([x,y])
print(xy_pairs)
```

```python
# to data frame
import pandas as pd
df = pd.DataFrame(xy_pairs, columns=['x','y'])
print(df)
```

```python
# save file
df.to_csv('./mycsv.csv', index=False)
print('saved')
```
